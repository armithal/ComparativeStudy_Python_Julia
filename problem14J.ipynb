```julia
#FINAL SCRIPT WITH FULL SCRIPT TIMESTAMP AND TIME STAMP AFTER LOADING THE INPUT DATA AND before printing THE OUTPUT
import DataStructures
@time begin
struct Info
    k
    kmers
end

function DeBruijn(p::Info)
    prefix = String[]
    suffix = String[]
    for i in p.kmers
        push!(prefix,i[1:end-1])
        push!(suffix, i[2:end])
    end    
    graph = DataStructures.SortedDict()
    for i in range(start=1,stop=length(prefix),step=1)
        if !haskey(graph, prefix[i])    
            graph[prefix[i]] = [suffix[i]]
        else
            push!(graph[prefix[i]], suffix[i])
        end
    end
    
#function Cycle(p::Info)
    vnode = String
    knode = String
    outDegree = DataStructures.DefaultDict(0)
    inDegree = DataStructures.DefaultDict(0)
    for (k,v) in graph
        outDegree[k] = length(v)
    end
    for (k,v) in graph
        for node in keys(graph)
            if node ∈ v
                inDegree[node] += 1
            end
        end
        for node in v
            if node ∉ keys(graph)
                inDegree[node] += 1 
            end
        end
    end
    for node in keys(inDegree)
        if node ∉ keys(outDegree)
            outDegree[node] = 0
        end
    end
    for node in keys(outDegree)
        if node ∉ keys(inDegree)
            inDegree[node] = 0
        end
    end
    for node in keys(inDegree)
        outgoing = outDegree[node]
        incoming = inDegree[node]
        if incoming != outgoing
            if outgoing > incoming
                vnode = node
            else
                knode = node
            end
        end  
    end
    if knode in keys(graph)
        push!(graph[knode], vnode)
    else
        graph[knode] = [vnode]
    end
        
    #basically now we enter the Eulerian Graph function in the python code, and then we will come back
    cycle = []
    keylist = [k for k in keys(graph)]
    traverse = rand(keylist,1)
    starts = traverse[end]
    while length(traverse) >=1
        if graph[starts] in values(graph)
            node = pop!(graph[starts])
            while node != starts
                push!(traverse, node)
                node = pop!(graph[node])
            end
            push!(traverse, node)
            for k in range(start = 1, stop= length(traverse), step = 1)
                if length(graph[traverse[end]]) >= 1
                    starts = traverse[end]
                elseif length(graph[traverse[end]]) == 0    
                    push!(cycle, pop!(traverse))
                end
            end
        end  
    end
    pop!(cycle)
    cycle = reverse(cycle)
    path1 = []
    path2 = []
    if cycle[end] == knode && cycle[1] == vnode
        path1 = cycle[2:end]
        path2 = cycle[1]
    else
        for i in range(start = 2, stop = length(cycle), step = 1)
            if cycle[i-1] == knode && cycle[i] == vnode
                path1 = cycle[1:i-1]
                path2= cycle[i:end]
            end
        end
    end
    eulerianPath = vcat(path2,path1)
    
    #converting list to string
    finalStr = String(eulerianPath[1])
    for i in range(start = 1, stop = length(eulerianPath)-1,step = 1 )
        kmer1 = eulerianPath[i]
        kmer2 = eulerianPath[i+1]
        if kmer1[2:end] == kmer2[1:end-1]
            finalStr = finalStr * kmer2[end]
        end
        
    end
    return finalStr
end




function main()
    open("rosalind_ba3h_762_7_dataset.txt") do f
        s = readline(f)
        k = (parse(Int64, s))
      inp = [] 
    
      # read till end of file
      while ! eof(f) 
         # read a new / next line for every iteration          
         s = readline(f)         
         push!(inp, s)
      end
      #println(length(inp))
      @time begin
      p = Info(k, inp)
      a = DeBruijn(p)
      end
      println(a)
      println(length(a))
    
    end
end
main()
end
```
