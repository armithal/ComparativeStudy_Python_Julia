```julia
#THIS IS THE FINAL OUTCOMEIKELIHOOD CODE- JULIA
import DataStructures

struct Infor
    observedString::String
    possibleObservableStates
    possibleHiddenStates
    transitionDict::DataStructures.DefaultOrderedDict
    emissionDict::DataStructures.DefaultOrderedDict
end

function outcomeLikelihood(p::Infor)
    #calculate and return the probability of the observed string being emitted by an HMM
    pathProb = []
    initialProbabilities = DataStructures.DefaultOrderedDict(0)
    for i in 1:length(p.possibleHiddenStates)
        initialProbabilities[p.possibleHiddenStates[i]] = 1/ length(p.possibleHiddenStates) * p.emissionDict[p.observedString[1] * p.possibleHiddenStates[i]]
    end
    push!(pathProb, initialProbabilities)
    for t in 2:length(p.observedString)
        newPathProbabilities = DataStructures.DefaultOrderedDict(0)
        for hiddenstate in p.possibleHiddenStates
            tempProbDict = DataStructures.DefaultOrderedDict(0)
            for (k,v) in pathProb[t-1]
                tempProbDict[k * hiddenstate] = v * p.transitionDict[k*hiddenstate] * p.emissionDict[p.observedString[t]*hiddenstate]
            end
            newPathProbabilities[hiddenstate] = sum(values(tempProbDict))
        end
        push!(pathProb, newPathProbabilities)
    end
    outcome = 0 
    for v in values(pathProb[end])
        outcome = outcome + v
    end
    return outcome
end

function main()
    input = []  
    open("rosalind_ba10d_762_8_dataset.txt") do f
    # read till end of file
      while ! eof(f) 
         # read a new / next line for every iteration          
         s = readline(f) 
         push!(input, s) 
      end
    end
    observedString = input[1]
    alphabets = input[3]
    E = [x for x in split(alphabets) if length(x) > 0]
    state = input[5]
    states = [x for x in split(state) if length(x) > 0]
    tempSpace = []
    start = 0
    for ln in input[5:end]
        if start == 1
            if startswith(ln, "---")
                start = 0 
                continue
            end
            push!(tempSpace, ln)
        end
        if startswith(ln, "---")
            start = 1
        end
    end

    transition = []
    for i in tempSpace
        r = i
        s = [x for x in split(r) if length(x) > 0]
        push!(transition, s)
    end
    transitionDict = DataStructures.DefaultOrderedDict(0)
    for i in 1:length(states)
        for j in 2:length(states)+1
            transitionDict[transition[1][i] * transition[j][1]] = parse(Float64, transition[i + 1][j])
        end
    end
    
    tempSpace2 = []
    start = 0
    for ln in input[7:end]
        if start == 1
            if startswith(ln, "---")
                start = 0
                continue
            end
            push!(tempSpace2, ln)
        end
        if startswith(ln, "---")
            start = 1
        end
    end
    emission = []
    for i in tempSpace2
        r = i
        s = [x for x in split(r) if length(x) >0]
        push!(emission, s)
    end
    emissionDict = DataStructures.DefaultOrderedDict(0)
    for i in 1:length(E)
        for j in 2:length(states)+1
            emissionDict[emission[1][i] * emission[j][1]] = parse(Float64, emission[j][i + 1])
        end
    end
    @time begin
    p = Infor(observedString, E, states, transitionDict, emissionDict)
    o = outcomeLikelihood(p)
    end
    println(o)
end
main()
```
